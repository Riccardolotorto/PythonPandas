2. visualizzare i nomi degli atleti italiani

```python
atleti[atleti["NOC"] == "Italy"][["Name"]]
```

3. controllare se gli atleti italiani sono di piÃ¹ o di meno degli atleti francesi

```python
print("ci sono piÃ¹ atleti italiani" if atleti[atleti["NOC"] == "Italy"]["NOC"].count() > atleti[atleti["NOC"] == "France"]["NOC"].count() else "ci sono piÃ¹ atleti francesi")
```

4. visualizzare tutti gli atleti che si chiamano Giovanni

```python
atleti[atleti["Name"].str.contains("Giovanni")][["Name"]]
```

5.  visualizzare gli atleti che si chiamano Alberto ma che non sono italiani

```python
atleti[(atleti["Name"].str.contains("Alberto")) & (atleti["NOC"] != "Italy")][["Name"]]
```

6.  visualizzare il numero di atleti della squadra di basket spagnola

```python
atleti[(atleti["NOC"] == "Spain") & (atleti["Discipline"] == "Basketball")]["Discipline"].count()
```

7. visualizzare i dati del dataframe ordinati in ordine alfabetico sulla disciplina

```python
atleti.sort_values(by="Discipline", ignore_index = True)
```

8. visualizzare gli atleti scandinavi

```python
atleti[(atleti["NOC"] == "Norway") | (atleti["NOC"] == "Sweden") | (atleti["NOC"] == "Denmark")][["Name"]]
```

9.  eliminare tutti gli atleti della Francia

```python
atleti.drop(atleti[atleti["NOC"] == "France"].index, inplace = True)
```
